```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#get the data

df2014 = pd.read_csv("2014_Financial_Data.csv")
df2015 = pd.read_csv("2015_Financial_Data.csv")
df2016 = pd.read_csv("2016_Financial_Data.csv")
df2017 = pd.read_csv("2017_Financial_Data.csv")
df2018 = pd.read_csv("2018_Financial_Data.csv")
```

```python
#merge data from different years. line up PRICE VAR column from each year.

df2014.columns = df2014.columns.str.replace('[', '')
df2014.columns = df2014.columns.str.replace(']', '')
df2014.columns = df2014.columns.str.replace('%', '')

df2015.columns = df2015.columns.str.replace('[', '')
df2015.columns = df2015.columns.str.replace(']', '')
df2015.columns = df2015.columns.str.replace('%', '')

df2016.columns = df2016.columns.str.replace('[', '')
df2016.columns = df2016.columns.str.replace(']', '')
df2016.columns = df2016.columns.str.replace('%', '')

df2017.columns = df2017.columns.str.replace('[', '')
df2017.columns = df2017.columns.str.replace(']', '')
df2017.columns = df2017.columns.str.replace('%', '')

df2018.columns = df2018.columns.str.replace('[', '')
df2018.columns = df2018.columns.str.replace(']', '')
df2018.columns = df2018.columns.str.replace('%', '')

#Rename column 
df2014 = df2014.rename(columns={"2015 PRICE VAR " : "Price var"})
df2015 = df2015.rename(columns={"2016 PRICE VAR " : "Price var"})
df2016 = df2016.rename(columns={"2017 PRICE VAR " : "Price var"})
df2017 = df2017.rename(columns={"2018 PRICE VAR " : "Price var"})
df2018 = df2018.rename(columns={"2019 PRICE VAR " : "Price var"})

#Concat datasets
df = pd.concat([df2014, df2015, df2016, df2017, df2018], axis = 0)
pd.set_option('display.max_columns', None)
```

```python
# drop columns with 10% null values and 20% zero values

temp_df = df["Class"]
df = df.drop(["Class"], axis="columns") 

df = df.loc[:, df.isnull().mean() < .1]
df = df.loc[:, (df==0).mean() < .2]

df = pd.concat([df,temp_df],axis=1)
```

```python
#drop column with name of company.

df = df.drop(["Unnamed: 0"], axis = "columns")
```

```python
#sector feature must be encoded to numerical

from sklearn.preprocessing import LabelEncoder

le_sector = LabelEncoder()
df["Sector_n"] = le_sector.fit_transform(df["Sector"])
encoded_df = df.drop(["Sector"], axis = "columns")
```

```python
#drop rows with nulls. Imputing doesnt make sense when all companies are completely different and we have noisy data.

cleaned_df = encoded_df.dropna()
```

```python
# prep for features normalization
target = cleaned_df["Class"]
features = cleaned_df.drop(["Class"], axis = "columns")
```

```python
# normalizing the numerical features

categorical_features = features["Sector_n"]
numerical_features = features.drop(["Sector_n"], axis = "columns")

from sklearn.preprocessing import StandardScaler
scaled_features = StandardScaler().fit_transform(numerical_features.values)

scaled_features_df = pd.DataFrame(scaled_features, index=numerical_features.index, columns=numerical_features.columns)

normalized_df = pd.concat([scaled_features_df, categorical_features, target], axis = 1)
normalized_df
```

```python
# clustering to remove outliers.

from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=3, min_samples=2).fit(scaled_features_df)
labels = clustering.labels_
df_labels = pd.DataFrame(labels, columns = ['Labels'])
normalized_df = normalized_df.reset_index()
final_df = pd.concat([normalized_df, df_labels], axis = 1)
```

```python
#Remove all clusters except 1 - which has 95% of the data.

final_df = final_df[final_df.Labels == 1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print("Linear Regression: ")
print("R2 is: ", lr.score(X_test, y_test))
print("mse is: ", mean_squared_error(y_pred, y_test))
print("mae is: ", mean_absolute_error(y_pred, y_test))
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_nb = gnb.predict(X_test)
print("Naive Bayes:")
print("Accuracy: ", accuracy_score(y_test, y_pred_nb))
print("F1 Score:: ", f1_score(y_test, y_pred_nb, average = "binary"))
```

```python
from sklearn import tree
from sklearn.model_selection import GridSearchCV

#different hyper paremeters

param_grid = {
    'max_depth': [5, 15, 25, 50],
    'min_samples_leaf': [1, 3, 6, 12],
    'max_leaf_nodes': [10, 20, 35, 50, 100]}

#tree and gridsearch, creating different trees with all combinations of parameters

dt = tree.DecisionTreeClassifier()
gs = GridSearchCV(dt, param_grid, scoring="accuracy", cv=5)
```

```python
#run gridsearch with different params on a dtree

gs.fit(X_train, y_train)
y_pred_dtree = gs.predict(X_test)
print("Decision Tree: ")
print("D_tree Best Parameters: ", gs.best_params_)
print("Accuracy: ", gs.score(X_test, y_test))
print("F1 Score: ", f1_score(y_test, y_pred_dtree, average = "binary"))
```

```python
#best features to split the tree on
bing = zip(X_train.columns.values,gs.best_estimator_.feature_importances_)
feature_importances_df = pd.DataFrame([{'colname':colname, 'importance':importance} for colname, importance in bing])
feature_importances_df = feature_importances_df.sort_values('importance', ascending=False).head(20)
feature_importances_df = feature_importances_df.set_index('colname')
feature_importances_df.plot.bar()
```

```python
#neural network

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'activation': ["relu"],#, "tanh", "logistic"],
    'solver': ["adam"],#, "sgd"],
    'alpha': [0.0001, 0.0005, 0.001],
    'learning_rate_init': [0.0001, 0.0005, 0.001, 0.005]}

mlp = MLPClassifier(max_iter=1000, hidden_layer_sizes=(100, 100))
gsnn = GridSearchCV(mlp, param_grid, scoring="accuracy")
```

```python
#run neural network with different alpha/learning rate

gsnn.fit(X_train, y_train)
y_pred_nn = gsnn.predict(X_test)
print("Neural Network: ")
print("Neural Network Parameters: ", gsnn.best_params_)
print("Accuracy: ", gsnn.score(X_test, y_test))
print("F1 Score: ", f1_score(y_test, y_pred_nn, average = "binary"))
```
